```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import  seaborn as sns
import  os
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('googleplaystore.csv',parse_dates=['Last Updated'])
```

```python
a=dataset.info()
```

```python
rename = ['app','category','rating','reviews','size','install','type','price','content_rating',
         'geners','last_update','cur_version','Android_ver']
dataset.columns=rename
```

```python
dataset.info()
```

```python
#dataset.dropna(axis=0, inplace=True)
```

```python
dataset['rating'] = dataset.rating.replace(19.0,1.9)
```

```python
dataset['category'] = dataset.category.str.replace('1.9','')
```

```python
dataset['install']= dataset.install.str.replace('+','')
```

```python
dataset['install']= dataset.install.str.replace(',','')
```

```python
dataset['install']= dataset.install.str.replace('Free','0')
```

```python
dataset['reviews']=dataset.reviews.str.replace('.0M','000000')
```

```python
dataset['reviews']=dataset['reviews'].astype(int)
```

```python
dataset['install']=dataset['install'].astype(int)
```

```python
scaller = StandardScaler()
```
